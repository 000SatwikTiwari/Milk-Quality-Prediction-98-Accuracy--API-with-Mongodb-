```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
import statistics as stat
from tkinter import *
import tkinter
from flask import Flask,request,jsonify
import pymongo
from pymongo import MongoClient
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("milknew.csv")
df
```

```python
df.isnull().sum()
```

```python
x=df[["pH","Temprature","Taste","Odor","Turbidity","Colour"]]
```

```python
y=df[["Grade"]]
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.30, random_state =250) #to maximize the accuracy
```

```python
logreg = LogisticRegression()
knn_classifier = KNeighborsClassifier(10)
nb=GaussianNB()
logreg.fit(xtrain, ytrain)
knn_classifier.fit(xtrain, ytrain)
nb.fit(xtrain, ytrain)
logreg_ypred = logreg.predict(xtest)
knn_classifier_ypred = knn_classifier.predict(xtest)
nb_ypred=nb.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score

logreg_acc = accuracy_score(ytest, logreg_ypred)
knn_classifier_acc = accuracy_score(ytest, knn_classifier_ypred)
nb_acc = accuracy_score(ytest, nb_ypred)
```

```python
print ("Logistic Regression accuracy : ", round(logreg_acc*100, 2))
print ("K-NN Classifier accuracy     : ", round(knn_classifier_acc*100, 2))
print ("Naive Biase accuracy       : ", round(nb_acc*100, 2))
```
